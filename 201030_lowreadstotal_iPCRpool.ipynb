```python
import pandas as pd
import os
import re
```

```python
path = os.listdir("/ghds/cv19/analysis")
r = re.compile(r"20(09|10|11).{27}$")
file_names2 = list(filter(r.match, path))
```

```python
low_reads = []
for i in file_names2:
    current = pd.read_csv(f"/ghds/cv19/analysis/{i}/pool_stats.hdr.tsv", sep='\t').iloc[:, :4]
    std_dev = current['reads_total'].std()
    mean = current['reads_total'].mean()
    lower_bound = mean - 2 * std_dev
    current['mean'] = mean
    current['lower_bound'] = lower_bound
    low_reads.append(current.query(f'reads_total < {lower_bound}'))
    
df2 = pd.concat(low_reads)
df2 = df2.sort_values(by='pos_pooling')

well_position = []
date = []
for _, value in df2['pos_pooling'].items():
    well_position.append(value[14:])

for _, value in df2['runid'].items():
    date.append(value[:6])
    
df2['well_position'] = well_position
df2['date'] = date
```

```python
flow_cell = df2['runid'].values
pos_pooling = df2['pos_pooling'].values

pos_extraction = []

for i,j in zip(flow_cell, pos_pooling):
    temp = pd.read_csv(f"/ghds/cv19/analysis/{i}/c19_read_counts.hdr.tsv", sep='\t')
    temp.dropna(inplace=True)
    pos_extraction.append(temp[temp['pos_pooling'] == j]['pos_extraction'].unique()[0][:13])

df2['Extraction_Plate_ID'] = pos_extraction

thermocycler_ID = ['TCC00142', 'TCC00174', 'TCC00040',
                   'TCC00154', 'TCC00104', 'TCC00104',
                   'TCC00115', 'TCC00076', 'TCC00121',
                   'TCC00102', 'TCC00102', 'TCC00121',
                   'TCC00115', 'TCC00142', 'TCC00172',
                   'TCC00060', 'TCC00144', 'TCC00149', 
                   'TCC00148', 'TCC00173', 'TCC00151', 
                   'TCC00175', 'TCC00142', 'TCC00178', 
                   'TCC00150', 'TCC00177', 'TCC00143']

df2['thermocycler_ID'] = thermocycler_ID
```
