```python
from IPython.display import display
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
import os
import re
```

```python
path = os.listdir('20201112_iPCR_troubleshoot_regionmolarity/')
r = re.compile(r".*\.csv$")
path = list(filter(r.match, path))
path.sort(reverse=True)
```

```python
outliers=[]
for i in path:
    temp = pd.read_csv(f'20201112_iPCR_troubleshoot_regionmolarity/{i}', encoding='unicode_escape')
    temp['FileName']=i
    zscore = np.abs(stats.zscore(temp['Region Molarity [nmol/l]']))
    outliers.append(temp.iloc[np.where(zscore > 3)])
```

```python
for i in outliers:
    display(i)
```

```python
total_cols = 2
total_rows = 3
fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                        figsize=(7*total_cols, 5*total_rows), constrained_layout=True, squeeze=False)
sns.set(font_scale = 1.1)

cbar_ax = fig.add_axes([.91, .3, .03, .4])

for i,k in zip(path, range(len(path))): 
    quant = pd.read_csv(f'20201112_iPCR_troubleshoot_regionmolarity/{i}', encoding='unicode_escape')

    row_position = []
    column_position = []

    for _, value in quant['WellId'].items():
        row_position.append(value[0])
        column_position.append(int(value[1:]))

    quant['row'] = row_position
    quant['column'] = column_position
    quant = quant.iloc[:,[7,10,11]].pivot(index='row', columns='column', values='Region Molarity [nmol/l]')
    
    row = k//total_cols
    pos = k % total_cols
    sns.heatmap(quant, annot=True, cmap="PuBu", fmt='.3g', vmin=0, annot_kws={"fontsize":12}, ax=axs[row][pos])
    axs[row][pos].set_title(f'Region Molarity [nmol/l] of Plate: {i}')
    
plt.show()
```
